```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
customers = pd.read_csv("../data/customers.csv")
employees = pd.read_csv("../data/employees.csv")
menu = pd.read_csv("../data/menu.csv")
orders = pd.read_csv("../data/orders.csv")
order_details = pd.read_csv("../data/order_details.csv")
```

```python
customers_df = customers.copy()
employees_df = employees.copy()
menu_df = menu.copy()
orders_df = orders.copy()
order_details_df = order_details.copy()
```

```python
for name, df in {
    "Customers": customers_df,
    "Employees": employees_df,
    "Menu": menu_df,
    "Orders": orders_df,
    "Order Details": order_details_df
}.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

```python
for name, df in {
    "Customers": customers_df,
    "Employees": employees_df,
    "Menu": menu_df,
    "Orders": orders_df,
    "Order Details": order_details_df
}.items():
    print(name, ":", df.duplicated().sum())
```

```python
customers_df.dtypes
orders_df.dtypes
menu_df.dtypes
```

```python
orders_df["Order_Date"] = pd.to_datetime(orders_df["Order_Date"])
```

```python
orders_df["Month"] = orders_df["Order_Date"].dt.month_name()
```

```python
orders_df["Day"] = orders_df["Order_Date"].dt.day_name()
```

```python
orders_df["Year"] = orders_df["Order_Date"].dt.year
```

```python
orders_df["Weekend"] = orders_df["Day"].isin(
    ["Saturday", "Sunday"]
)
```

```python
sales = orders_df.merge(
    order_details_df,
    on="Order_ID",
    how="inner"
)
```

```python
sales = sales.merge(
    menu_df,
    on="Item_ID",
    how="left"
)
```

```python
sales = sales.merge(
    customers_df,
    on="Customer_ID",
    how="left"
)
```

```python
sales = sales.merge(
    employees_df,
    on="Employee_ID",
    how="left"
)
```

```python
sales.head()
```

```python
#revenue
sales["Revenue"] = (
    sales["Selling_Price"] *
    sales["Quantity"]
)
```

```python
#cost
sales["Cost"] = (
    sales["Cost_Price"] *
    sales["Quantity"]
)
```

```python
#profit
sales["Profit"] = (
    sales["Revenue"] -
    sales["Cost"]
)
```

```python
sales.to_csv(
    "../data/cleaned_sales.csv",
    index=False
)
```
